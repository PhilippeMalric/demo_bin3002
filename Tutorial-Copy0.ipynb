La première étape est de créer la base de données correspondant à ce schéma. Pour ce faire, il est nécessaire d'utiliser la commande SQL ``CREATE``. Le site du cours présente des exemples.
```python
def create_tables():
    """Execute les commandes CREATE afin de generer le schema presente plus tot."""
    # Etablir une connexion.
    # Attention! Sur esilbac, host="esilbac2"
    con = pymysql.connect(
        host="",
        user="p0947411",
        passwd="patate", # Ceci n'est pas mon mot de passe sur esilbac...
        db="p0947411_bin3002",
    )
    
    # Obtenir un curseur. C'est l'objet sur lequel on peut appeller la methode execute
    # afin d'executer des requetes SQL.
    cur = con.cursor()
    
    sql = """
        CREATE TABLE Gene (
          ensemblid VARCHAR(45) NOT NULL,
          symbol VARCHAR(45) NULL,
          chr VARCHAR(2) NOT NULL,
          geneStart INT NOT NULL,
          geneEnd INT NOT NULL,
          strand VARCHAR(1) NOT NULL,
          PRIMARY KEY (ensemblid)
        )
    """ # Cette requete cree la table Gene.
    # Vous devrez utiliser une syntaxe similaire afin de creer chaque table.
    
    cur.execute(sql)
    
    # Je ne commenterai pas le reste, mais il s'agit simplement de creer les tables.
    # Il faut garder en tete que si A.b est une cle etrangere vers la table B, il faudra
    # creer la table B en premier.
    sql = """
        CREATE TABLE Locus (
          id INT NOT NULL AUTO_INCREMENT,
          chr VARCHAR(2) NOT NULL,
          pos INT NOT NULL,
          gene VARCHAR(45) NULL,
          CONSTRAINT fk_locus_gene FOREIGN KEY (gene) REFERENCES Gene(ensemblid),
          PRIMARY KEY (id)
        )
    """ # Remarquez la creation d'une cle etrangere.
    cur.execute(sql)

    sql = """
        CREATE TABLE Variant (
          id INT NOT NULL AUTO_INCREMENT,
          locus INT NOT NULL,
          rs VARCHAR(45) NULL,
          ref VARCHAR(1) NOT NULL,
          alt VARCHAR(1) NOT NULL,
          CONSTRAINT fk_var_locus FOREIGN KEY (locus) REFERENCES Locus(id),
          PRIMARY KEY (id)
        )
    """
    cur.execute(sql)

    sql = """
        CREATE TABLE Genotype (
          variant INT NOT NULL,
          sampleid VARCHAR(45) NOT NULL,
          geno INT NOT NULL,
          CONSTRAINT fk_geno_var FOREIGN KEY (variant) REFERENCES Variant(id),
          CONSTRAINT pk_geno PRIMARY KEY (variant, sampleid)
        )
    """
    cur.execute(sql)

    sql = """
        CREATE TABLE Expression (
          gene VARCHAR(45) NOT NULL,
          sampleid VARCHAR(45) NOT NULL,
          value FLOAT NOT NULL,
          CONSTRAINT fk_exp_gene FOREIGN KEY (gene) REFERENCES Gene(ensemblid),
          CONSTRAINT pk_expr PRIMARY KEY (gene, sampleid)
        )
    """
    cur.execute(sql)
    
    con.commit() # S'assure que tout a ete execute.
    con.close() # Ferme la connexion.
    
# create_tables() # Il faut executer cette fonction seulement une fois.
```

```python
def fill_gene():
    """Remplit la table Gene en utilisant des donnees issues de Ensembl (voir TP1). """
    
    con = pymysql.connect("", "p0947411", "patate", "p0947411_bin3002")
    cur = con.cursor()
    
    # On construit le chemin vers le fichier de donnees.
    fn = os.path.join("data", "ensembl_genes_filtered.txt")
    
    # On peut ensuite lire le fichier ligne par ligne.
    with open(fn) as f:
        for line in f:
            # Ici line sera un object 'str' contenant chaque ligne du fichier.
            # Je retire le retour chariot.
            line = line.rstrip("\n")
            
            # Je convertit le 'str' en 'list' ou chaque colonne 
            # separee par un espace devient un element de la liste.
            line = line.split("\t")
            # A titre d'exemple:
            # print(line)
            # ['ENSG00000215405', '20737094', '-1', '20747114', '15', 'GOLGA6L6']
            
            # Il ne me reste plus qu'a convertir les types et a inserer
            # les elements dans la table.
            
            ensemblid, start, strand, end, chrom, symbol = line
            start = int(start)
            end = int(end)
            # J'encode le strand en + ou -.
            strand = '+' if strand == '1' else '-'
            cur.execute(
                "INSERT INTO Gene VALUES(%s, %s, %s, %s, %s, %s)",
                (ensemblid, symbol, chrom, start, end, strand)
            )
            
    con.commit()
    con.close()

# fill_gene() # Executer seulement une fois.
```

Maintenant, il faut remplir toutes les autres tables. Je commencerai avec Locus, Variant et Génotype à partir du fichier fourni la semaine du TP1.

```python
def fill_genetic_info():
    """Remplit les tables Locus, Variant et Genotype en fonction du fichier de genotypes Leprosy.genotypes.dat"""
    
    con = pymysql.connect("", "p0947411", "patate", "p0947411_bin3002")
    cur = con.cursor()
    
    fn = os.path.join("data", "Leprosy.genotypes.dat")
    # Comme il y a beaucoup de variants dans le fichier et que ceci est un exemple,
    # je vais seulement remplir les 10 000 premieres variations.
    num_snp = 0
    samples = None
    with open(fn) as f:
        for line in f:
            line = line.rstrip("\n")
            line = line.split("\t")
            if samples is None:
                samples = line[6:]
            else:
                num_snp += 1
                rs, chrom, pos, ref_homo, hetero, alt_homo = line[:6]
                pos = int(pos)
                # D'abord, il faut chercher dans quel gene se trouve ce variant.
                cur.execute(
                    "SELECT ensemblid FROM Gene WHERE chr=%s AND %s>geneStart AND %s<geneEnd",
                    (chrom, pos, pos)
                )
                gene = cur.fetchone() # Prend le premier tuple retourne par la requete.
                
                # Il faut creer le locus et le variant.
                cur.execute(
                    "INSERT INTO Locus (chr, pos, gene) VALUES (%s, %s, %s)",
                    (chrom, pos, gene)
                )
                locus = cur.lastrowid # Noter que cur.lastrowid contient le ID du dernier insert.
                
                cur.execute(
                    "INSERT INTO Variant (locus, rs, ref, alt) VALUES (%s, %s, %s, %s)",
                    (locus, rs, ref_homo[0], alt_homo[0])
                )
                variant = cur.lastrowid
                
                # Il faut maintenant remplir la table de Genotype qui contient l'information
                # sur tous les echantillons.
                samples_geno = line[6:]
                # Le assert est pour s'assurer que nous avons le meme nombre de ID dans notre header
                # que dans chaque rangee (precaution).
                assert len(samples) == len(samples_geno)
                
                # enumerate prend une liste et renvoie un iterateur de l'index et de la valeur.
                # Vous comprendrez mieux en experimentant avec cet outil tres utile.
                for i, sample in enumerate(samples):
                    geno = float(samples_geno[i])
                    cur.execute(
                        "INSERT INTO Genotype VALUES (%s, %s, %s)",
                        (variant, sample, geno)
                    )
            
            # Arreter de remplir apres 1000 entrees.
            if num_snp >= 10000:        
                con.commit()
                con.close()
                return
                
# fill_genetic_info()
```

```python
def fill_expression():
    """Remplit la table Expression a l'aide de Leprosy.expression.dat """
    
    con = pymysql.connect("", "p0947411", "patate", "p0947411_bin3002")
    cur = con.cursor()
    
    fn = os.path.join("data", "Leprosy.expression.dat")
    samples = None
    with open(fn) as f:
        for line in f:
            line = line.rstrip("\n")
            line = line.split("\t")
            if samples is None:
                samples = line[2:]
            else:
                ensembl_id, hugo_id = line[:2]
                # Conversion des valeurs d'expressions en float. 
                # J'ai utilise une "list comprehension" 
                # https://docs.python.org/3/tutorial/datastructures.html?highlight=list%20comprehension#list-comprehensions
                expression = [float(i) for i in line[2:]]
                
                # Il faut verifier si le gene est dans la BDD.
                cur.execute(
                    "SELECT ensemblid FROM Gene WHERE ensemblid=%s",
                    (ensembl_id)
                )
                if cur.fetchone() is not None:
                    for i, e in enumerate(expression):
                        cur.execute(
                            "INSERT INTO Expression VALUES (%s, %s, %s)",
                            (ensembl_id, samples[i], e)
                        )
    
    con.commit()
    con.close()
    
# fill_expression()
```

Maintenant, j'ai rempli toutes les tables avec l'information demandée. Il ne reste plus qu'à faire quelques requêtes qu'on peut effectuer à travers Python ou un terminal MySQL.

```python
con = pymysql.connect("", "p0947411", "patate", "p0947411_bin3002")
cur = con.cursor()

# J'importe deux librairies tres utiles en bio-informatique.
# Il s'agit d'un usage plus avance que ce que vous aurez a faire dans
# le TP, mais on peut voir pourquoi une base de donnees pourrait
# etre utile.
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

# Tracons l'histogramme des valeurs d'expression.
# Extraire l'information de la BD. 
cur.execute("SELECT value FROM Expression")
exp = np.array(cur.fetchall())
plt.hist(exp, facecolor="black", edgecolor="white", bins=15)
plt.xlabel("Expression")
plt.show()

# Trouvons les variants du gene ENSG00000116688.
# Noter les conditions de jointure. 
# La jointure permet de "fusionner" les tuples issus de deux
# tables. Par exemple, ici la table Variant a une cle etrangere
# (le champ locus) qui "pointe" vers la cle primaire (id) de la
# table Locus. Or, la condition de jointure `Variant.locus=Locus.id``
# forme des tuples de la forme suivante:
#
# Locus (id, chr, pos, gene)
# Variant (id, locus (FK), rs, ref, alt)
# Locus JOIN Variant:
# (Locus.id = Variant.locus, Locus.chr, Locus.pos, Locus.gene, Variant.id,
#  Variant.rs, Variant.ref, Variant.alt)

cur.execute(
    "SELECT Gene.symbol, Variant.rs "
    "FROM Variant, Locus, Gene "
    "WHERE Variant.locus=Locus.id "
    "  AND Locus.gene=Gene.ensemblid "
    "  AND Gene.ensemblid='ENSG00000116688'"
)
for gene, rs in cur.fetchall():
    print(gene, rs)
print("--------------\n")

# Nous allons maintenant selectionner les valeurs d'expression les 10 plus petites
# pour le gene ENSG00000116688
cur.execute(
    "SELECT e.sampleid, e.value "
    "FROM Expression e, Gene g "
    "WHERE e.gene=g.ensemblid " # Join
    "  AND g.ensemblid='ENSG00000116688' " # Selection du gene
    "  ORDER BY e.value ASC " # Tri ascendant sur la valeur de l'expression
    "LIMIT 5" # Sortir seulement les 5 premiers resultats
)
for sample, expression_val in cur.fetchall():
    print(sample, expression_val)
    
con.close()
```
